# 5. Analisi di un dataset di video giochi
1. Quanti videogiochi sono stati pubblicati
2. Quali sono i generi più popolari? Mostralo con un bar plot
3. Plotta l'evoluzione del numero di giochi pubblicati nel tempo
4. Prova a riprodurre questo plot 



```python
# Import dei pacchetti
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento del dataset
url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)
df.head()
```

```python
#quanti giochi sono stati pubblicati
len(df)
```

```python
#barplot generi più popolari
popularity = df.groupby('Genre')['Global_Sales'].sum()
plt.figure(figsize=(12,10))
plt.bar(x=popularity.index, height=popularity.values, color='green')
plt.title('Generi più popolari')
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#evoluzione del numero di giochi pubblicati nel tempo
n_games = df['Year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.plot(n_games.index, n_games.values, marker='o')
plt.title('Evoluzione numero di giochi per anno')
plt.xlabel('Anno')
plt.ylabel('Numero di giochi')
plt.grid()
plt.show()
```

```python
#riproduzione plot
df_filtrato = df[df['Genre'].isin(['Action', 'Misc', 'Role-Playing', 'Shooter', 'Sports'])]
categorie = df_filtrato.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
colors = ['Blue', 'Orange', 'Green', 'Red']
base = None

for regione, colore in zip(categorie.columns, colors):
    plt.bar(categorie.index, categorie[regione], bottom=base, label=regione,color=colore)
    if base is None:
        base = categorie[regione]
    else:
        base = base + categorie[regione]

plt.title('Vendite per regione per genere')
plt.xlabel('Genre')
plt.ylabel('Milioni di unità')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

#df['Action', 'Misc', 'Role-Playing', 'Shooter', 'Sports']
```
