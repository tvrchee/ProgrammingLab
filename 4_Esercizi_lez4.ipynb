## 1. La legge dei grandi numeri

La legge dei grandi numeri è un teorema fondamentale della teoria della probabilità che indica che se ripetiamo molte volte (tendendo all'infinito) lo stesso esperimento, la frequenza di un certo evento tende ad essere costante.

Vale a dire, **la legge dei grandi numeri indica che se lo stesso test viene eseguito ripetutamente (ad esempio, lancio di una moneta), la frequenza con cui si ripeterà un determinato evento (testa/croce) si avvicinerà a una costante. Questa a sua volta sarà la probabilità che questo evento si verifichi.**

Fonte: https://it.economy-pedia.com/11038376-law-of-the-big-numbers

- Scrivere una funzione che simula il lancio di una moneta per N volte e restituisce la frequenza % di volte che è uscita testa.
- Calcolare la frequenza % con la grandezze del campione (numero di lanci) che varia 100 volte da 10 a 20000. Ottengo quindi 100 valori.
- Fare il plot del Numero medio di teste (y) rispetto alla grandezza del campione (x)





```python
import numpy as np
import matplotlib.pyplot as plt

def teste(n_lanci):
    lanci = [np.random.randint(0, 2) for i in range(n_lanci)]
    n_teste = sum(lanci)
    frequenza = (n_teste/n_lanci)*100
    print(f'In {n_lanci} abbiamo ottenuto {n_teste} teste, quindi con frequenza: {frequenza}%')

teste(100)
```

```python
grandezza_campione = np.arange(10, 20000, 100)
lista_frequenze = [(sum(np.random.randint(0, 2) for i in range(size))/size) for size in grandezza_campione] #per ogni 'size' (num di lanci) calcola la frequenza relativa di teste
#in pratica lancia la moneta 'size' volte e somma quante volte è uscito 1(testa) da estrazione casuale, poi divide per numero totale di lanci

lista_frequenze
```

```python
plt.figure(figsize=(12,6))
plt.axhline(y = 0.5, color = 'orange', linewidth = 3)
plt.plot(grandezza_campione, lista_frequenze, '-o', color = 'skyblue')
plt.title('Numero medio di teste rispetto alla grandezza del campione')
plt.xlabel('Grandezza campione')
plt.ylabel('Numero medio teste')
plt.show()
```

## 2. Analisi del Dataset Titanic
- Quante righe e colonne ha il dataset?
- Controlla quanti valori mancanti ci sono per colonna
- Riempi i valori mancanti nella colonna 'Embarked' con il valore più frequente
- Rimuovi le righe dove il valore di 'Age' è mancante
- Controlla se ci sono righe duplicate
- Calcola l'età media dei passeggeri per ogni classe (`Pclass`), se ci sono valori mancanti di età riempili con il valore medio
- Visualizza la distribuzione dell'età per classe
- Visualizza la distribuzione dell'età per classe dividendo e mmostrando insieme i dati di uomini e donne

```python
# 📦 Caricamento del dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
len(df.columns) #numero colonne
len(df) #numero righe

for col in df.columns: #valori mancanti per colonna
    mancanti = len(df) - sum(df[col].notna())
    print(f"Colonna {col}: {mancanti} valori mancanti\n")

moda_embarked = df["Embarked"].mode()[0] #riempio i valori NA con il + frequente
df['Embarked'].fillna(moda_embarked, inplace=True) 

df.dropna(subset=['Age'], inplace=True) #rimuovo righe con valori 'Age' mancanti 

print(f"\nRighe duplicate: {len(df) - len(df.drop_duplicates(inplace=False))}") #controllo se ci sono righe duplicate 
df.drop_duplicates() #rimuovo le righe duplicate

df['Age'].fillna(df['Age'].mean(), inplace=True)
mean_ages = df.groupby('Pclass')['Age'].mean() #calcolo età media per classe di passeggeri 
print(f"Età media per classe di passeggeri: {mean_ages}")

sns.boxplot(data=df, x='Pclass', y='Age', hue="Pclass")
plt.title("Distribuzione dell'età per classe")
plt.show()

sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.title("Distribuzione dell'età per classe dividendo uomini e donne")
plt.show()
```

## 3. Analisi del Dataset Iris

- Esplora la distribuzione delle specie (conta quanti campioni ci sono per specie)
- Calcola la lunghezza e la larghezza media dei petali per specie
- Visualizza le dimensioni dei petali per specie (scatterplot)
- Crea una nuova colonna per l'area del petalo e analizzala
- Grafico della distribuzione dell'area del petalo per specie (boxplot)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df.head()
```

```python
#primo punto
print(f"Distribuzione delle specie: {df['species'].value_counts()}")

#secondo punto
medie_petali_per_specie = df.groupby('species')[['petal_length', 'petal_width']].mean() 

#terzo punto
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
plt.figure(figsize=(10,6))
for specie in df['species'].unique():
    dati_specie = df[df['species'] == specie]
    plt.scatter(dati_specie['petal_width'], dati_specie['petal_length'], color=colors[specie], label=specie)
plt.title('Dimensioni dei petali per specie')
plt.xlabel('Larghezza petali per specie')
plt.ylabel('Lunghezza petali per specie')
plt.grid(True)
plt.show()

#quarto punto
df['petal_area'] = (df['petal_length'].astype(int)*df['petal_width'].astype(int))

#quinto punto
sns.boxplot(data=df, x='species', y='petal_area', hue='species')
plt.title("Distribuzione dell'area del petalo per specie")
plt.show()
```
